```python
from mpl_toolkits import mplot3d
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("../input/mcdonalds-india-menu-nutrition-facts/India_Menu.csv")
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,3))
sns.lmplot(x="Energy (kCal)",y="Protein (g)",data=df,hue="Menu Category")
```

```python
plt.figure(figsize=(12,3))
sns.jointplot(x="Total carbohydrate (g)", y="Total Sugars (g)",
              data=df,kind="hex", color="#00008B")
```

```python
df.corr()
```

```python
plt.figure(figsize=(16,8))
mask=np.zeros_like(df.corr())
chng=np.triu_indices_from(mask)
mask[chng]=True
rj=sns.heatmap(df.corr(),mask=mask,annot=True,cmap="GnBu",annot_kws={'size':16})
rj.set_title("Correlation Heat Map",fontdict={'fontsize':20})
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
